1: Handling missing values
There are 5 ways of dealing with missing values:

- mean()     filling with the average of the column         df[col].fillna(df[col].mean(), inplace=True)
- median()   filling with the middle of the column          df[col].fillna(df[col].median(), inplace=True)
- mode()     filling with the most frequent of the column   df[col].fillna(df[col].mode()[0], inplace=True)
- fixed()    filling with fixed values                      df[col].fillna(value, inplace=True)
- drop()     dropping rows or columns                       df.dropna(inplace=true) or df.drop('col', inplace=True)
