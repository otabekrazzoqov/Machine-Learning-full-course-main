```python
import pandas as pd

df_encoded = pd.read_csv('C:/Users/mrcoo/ml_course/ML-course/lesson_13/ai_assistant_usage_student_life.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
def tahrir(df_encoded):
    for col in df_encoded.select_dtypes(include=['object']).columns:
       cardinality=df_encoded[col].nunique()
       if cardinality<=5:
           le=LabelEncoder()
           df_encoded[col]=le.fit_transform(df_encoded[col])
       else:
        df_encoded=pd.get_dummies(df_encoded,columns=[col],dtype=int, drop_first=True)
    return(df_encoded)
```
