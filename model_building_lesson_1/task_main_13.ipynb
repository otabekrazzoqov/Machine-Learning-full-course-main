### filling missing values


```python
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df
```

### ORDINAL ENCODING

```python
custom_mapping_dict = {
    'Undergraduate':0,
    'High School':1,
    'Graduate':2}

df['StudentLevel_Ordinal'] = df['StudentLevel'].map(custom_mapping_dict)

df.drop(columns='StudentLevel', axis=1, inplace=True)
```

### FREQUENCY ENCODING

```python
df['Discipline'].value_counts()
```

```python
df['Discipline_frequency'] = df['Discipline'].map(df['Discipline'].value_counts())
df['TaskType_frequency'] = df['TaskType'].map(df['TaskType'].value_counts())
df['FinalOutcome_frequency'] = df['FinalOutcome'].map(df['FinalOutcome'].value_counts())
```

```python
df.drop(columns='Discipline', axis=1, inplace=True)
df.drop(columns='TaskType', axis=1, inplace=True)
df.drop(columns='FinalOutcome', axis=1, inplace=True)
```
