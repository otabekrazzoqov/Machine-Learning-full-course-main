```python
import pandas as pd
```

```python
df=pd.read_csv('train.csv')
```

```python
df.info()
```

# One - Hot encoding

```python
df=pd.get_dummies(df,columns=['Gender'],prefix='Gender',dtype=int)
```

```python
df.info()
```

# LabelEncoder

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
```

```python
df['Name']=encoder.fit_transform(df['Name'])
```

```python
df.info()
```

# Frequnce Encoding

```python
df.info()
```

```python
freq_encoding=df['Ticket'].value_counts()
df['Ticket_freq']=df['Ticket'].map(freq_encoding)
```

```python
df['Cabin_freq']=df['Cabin'].map(df['Cabin'].value_counts())
```

```python
df.info()
```

```python
df.drop('Ticket',axis=1,inplace=True)
```

```python
df.info()
```

```python
df['Parch'].value_counts()
```

# Target Encoding

```python
df.info()
```

```python
target_encoding = df.groupby('Name')['Survived'].mean()
df['Name_target'] = df['Name'].map(target_encoding)
```

# Ordinal Encoding

```python
df.info()
```

```python
df['Degree'].value_counts()
```

```python
degree_mapping={
    'Bachelor':0,
    'Master':1,
    'PhD':2
}

df['degree_ordinal']=df['Degree'].map(degree_mapping)
```

```python
df.info()
```

```python
df.drop('Degree',axis=1,inplace=True)
```

```python
df.info()
```

```python
df.info()
```

```python
df.columns.tolist()
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)


```
