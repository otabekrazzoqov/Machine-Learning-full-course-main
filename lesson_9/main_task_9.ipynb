```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv('C:/Users/mrcoo/ml_course/ML-course/lesson_9/us_house_Sales_data.csv')
```

### 

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df.sample(5)
```

## One-hot encoding

```python
df=pd.get_dummies(df,columns=['Bathrooms'],prefix='bath_num',dtype=int)
```

```python
df.columns
```

## Label Encoding

```python
encoder = LabelEncoder()
encoder
```

```python
df['Status'] = encoder.fit_transform(df['Status'])
```

```python
df.info()
```

## Frequence Encoding

```python
df.sample()
```

```python
df['Bedroom_freq']=df['Bedrooms'].map(df['Bedrooms'].value_counts())
```

```python
df.info()
```

```python
df.sample(10)
```

```python
df.Bedroom_freq.value_counts()
```
