```python
import pandas as pd
```

```python
df = pd.read_csv('job_salary_prediction_dataset.csv')
```

```python
df.info()
```

    Hangling missing values

    There are no missing values in this dataset so we do not have to worry about that 

    Encoding

```python
unique_counts = {}

for col in df.columns:
    unique_counts[col] = df[col].nunique()
print(unique_counts)
```

```python
df['education_level'].value_counts()
```

```python
education_mapping = {
    "High School" : 0,
    "Diploma" : 1,
    "Bachelor" : 2,
    "Master" : 3,
    "PhD" : 4
}

df['education_level'] = df['education_level'].map(education_mapping)
```

```python
df['company_size'].value_counts()
```

```python
company_mapping = {
    "Startup" : 0,
    "Small" : 1,
    "Medium" : 2,
    "Large" : 3,
    "Enterprise" :4
}

df['company_size'] = df['company_size'].map(company_mapping)
```

```python
df['remote_work'].value_counts()
```

```python
remote_work_mapping = {
    "No" : 0,
    "Hybrid" : 1,
    "Yes" :2
}

df['remote_work'] = df['remote_work'].map(remote_work_mapping)
```

    education_level, company_size and remote_works   columns are encoded with ordinal encoding

```python
df.info()
```

```python
cols = ["job_title", "industry", "location"]

from sklearn.preprocessing import LabelEncoder


for col in df.columns:
    if col in cols:
        encoder = LabelEncoder()
        df[col] = encoder.fit_transform(df[col])
 
```

```python
df.info()
```

    There are no visible outliers which means i don't have to scale columns.

```python
df.sample(10)
```

    1. LINEAR REGRESSION 

```python
x = df.drop("salary", axis=1)
y = df['salary']

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state = 42)

from sklearn.linear_model import LinearRegression
linear_rm = LinearRegression()

linear_rm.fit(x_train, y_train)
y_pred = linear_rm.predict(x_test)

from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

print('r2:',r2)
print('mae:',mae)
print('mse:',mse)
print('rmse:',rmse)
```

    DECISION TREE REGRESSOR

```python
from sklearn.tree import DecisionTreeRegressor
dt_m = DecisionTreeRegressor()

dt_m.fit(x_train, y_train)

y_pred = dt_m.predict(x_test)


from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
import numpy as np

r2=r2_score(y_test,y_pred)
mae=mean_absolute_error(y_test,y_pred)
mse=mean_squared_error(y_test,y_pred)
rmse=np.sqrt(mse)

print('r2:',r2)
print('mae:',mae)
print('mse:',mse)
print('rmse:',rmse)

```

```python
x = df.drop("education_level", axis=1)
y = df["education_level"]

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =0.2, random_state =42)

from sklearn.linear_model import LogisticRegression

lr = LogisticRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

from sklearn.metrics import classification_report

cl = classification_report(y_test, y_pred)

print(cl)
```

    DECISION TREE CLASSIFIER

```python
from sklearn.tree import DecisionTreeClassifier

dtc = DecisionTreeClassifier()

dtc.fit(x_train, y_train)

y_pred = dtc.predict(x_test)


from sklearn.metrics import classification_report

cl = classification_report(y_test, y_pred)

print(cl)
```
