```python
import pandas as pd
df=pd.read_csv('house_price.txt')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype=='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
def toldir(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype=='object':
                df[col].fillna(df[col].mode()[0],inplace=True)
            else:
                df[col].fillna(df[col].mean(),inplace=True)
    return df

```

```python
toldir(df)
```
