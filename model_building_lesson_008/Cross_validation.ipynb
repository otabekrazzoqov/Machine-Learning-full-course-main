### **Handling missing values**

### **Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
def encoding(df):
    
    for col in df.columns:
        if df[col].dtype == "object":
            if df[col].nunique() <= 3:
                dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                encoder = OrdinalEncoder()
                df[col] = encoder.fit_transform(df[[col]])
    return df
```

```python
df =  encoding(df)
df
```

```python
df.info()
```

### **Scaling**

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
df_scale = df.select_dtypes(include = ['float64','int64', 'int32']).columns.drop('sleep_disorder_risk')
df[df_scale] = scaler.fit_transform(df[df_scale])
```

```python
df.sample(10)
```

### **Logistic Regression**

```python
x = df.drop('sleep_disorder_risk', axis=1)
y = df['sleep_disorder_risk']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression(max_iter=1000)
lr
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import classification_report, mean_absolute_error
cl = classification_report(y_test, y_pred)
print(cl)
```

```python
mae = mean_absolute_error(y_test, y_pred)
```

### **Cross Validation**

```python
from sklearn.model_selection import KFold, cross_val_score

kf = KFold(n_splits=5, shuffle=True, random_state=42)
kf

scores = cross_val_score(lr, x,y,cv=kf, scoring="r2")
print(scores.mean())
```

### **MAE Cross Validation**

```python
from sklearn.metrics import make_scorer
mae=make_scorer(mean_absolute_error,greater_is_better=False)
scores=cross_val_score(lr,x,y,cv=kf,scoring=mae)
print(-scores)
```

### **Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
dtr.fit(x_train, y_train)
y_pred = dtr.predict(x_test)

```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np 
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)
```

```python
print(mae)
print(mse)
print(rmse)
print(r2)
```
