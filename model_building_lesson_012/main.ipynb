## **Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
encoder = OrdinalEncoder(categories=[["High School", "Undergraduate", "Graduate"]])
df["Academic_Level_encoded"] = encoder.fit_transform(df[["Academic_Level"]]).astype(int)
```

```python
df["Overall_Impact"].value_counts()
```

```python
encoder = OrdinalEncoder(categories=[["Negative", "Neutral", "Positive"]])
df['Overall_Impact_encoded'] = encoder.fit_transform(df[["Overall_Impact"]]).astype(int)
```

```python
df.drop(columns=["Academic_Level", "Overall_Impact"], inplace=True)
```

```python
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder
def encoding(df):
    encoder = LabelEncoder()
    for col in df.columns:
        if df[col].dtype=="object":
            if df[col].nunique() <=5:
                dummies = pd.get_dummies(df[col], prefix=col, dtype=int)
                df = pd.concat([df.drop(columns=[col]),dummies], axis=1)
            else:
                df[col] = encoder.fit_transform(df[col])
    return df
```

```python
df = encoding(df)
```

```python
df.info()
```

```python
df.sample(10)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
x = df.drop("Overall_Impact_encoded", axis=1)
y = df["Overall_Impact_encoded"]
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

rfc = RandomForestClassifier()

rfc.fit(x_train, y_train)
```

```python
y_pred = rfc.predict(x_test)
```

```python
from sklearn.metrics import classification_report

cl_r = classification_report(y_test, y_pred)

print(cl_r)
```
