| Code             | Purpose                      | Works On           | Notes                             |
| ---------------- | ---------------------------- | ------------------ | --------------------------------- |
| `fillna(mean)`   | Replace missing with average | int, float         | Sensitive to outliers             |
| `fillna(mode)`   | Replace with most frequent   | object, int, float | Good for categories               |
| `fillna(median)` | Replace with middle value    | int, float         | Robust to outliers                |
| `fillna(value)`  | Replace with constant        | any                | Manual default fill               |
| `drop(axis=1)`   | Drop column                  | column name        | Removes unwanted or empty columns |
| `dropna()`       | Drop rows with missing       | entire row         | Use carefully on large datasets   |


### Filling missing values

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype =='object':
            df[col].fillna(df[col].mode()[0],inplace=True)
        else:
            df[col].fillna(df[col].mean(),inplace=True)
```

| Encoding Type          | Description                                                          | Use Case Scenarios                                                                      | Pros                                                       | Cons                                                                                             |
| ---------------------- | -------------------------------------------------------------------- | --------------------------------------------------------------------------------------- | ---------------------------------------------------------- | ------------------------------------------------------------------------------------------------ |
| **One-Hot Encoding**   | Converts categories into binary vectors (one 1 and the rest 0)       | - Nominal variables (no order)<br>- Low cardinality (few unique values)                 | - Simple<br>- No ordinal assumptions<br>- Widely supported | - High dimensionality if many categories<br>- Not suitable for tree models without preprocessing |
| **Label Encoding**     | Assigns each category a unique integer                               | - Ordinal variables with true order<br>- Tree-based models                              | - Fast<br>- Memory-efficient                               | - Implies ordinal relationships even if not present<br>- Misleading for linear models            |
| **Frequency Encoding** | Replaces category with frequency of its occurrence                   | - High-cardinality features<br>- Tree-based models                                      | - Simple<br>- Preserves distribution information           | - Can introduce noise if frequencies are similar<br>- Not interpretable                          |
| **Ordinal Encoding**   | Assigns integers based on meaningful order of categories             | - Ordinal features (e.g. low, medium, high)                                             | - Captures order<br>- Simple                               | - Not suitable for nominal features<br>- Assumes linearity of order                              |
| **Target Encoding**    | Replaces category with the mean of target variable for that category | - High cardinality<br>- Useful in classification/regression (target leakage controlled) | - Captures target signal<br>- Compact                      | - Can cause overfitting<br>- Must be regularized or cross-validated                              |

# SCALING

| **Scaler**         | **What it Does**                                                            | **Formula**                                                       | **Sensitive to Outliers?** | **When to Use**                                                               |
| ------------------ | --------------------------------------------------------------------------- | ----------------------------------------------------------------- | -------------------------- | ----------------------------------------------------------------------------- |
| **StandardScaler** | Standardizes features by removing the mean and scaling to unit variance     | $z = \frac{x - \mu}{\sigma}$                                      | Yes                        | When data is normally distributed or has no extreme outliers                  |
| **MinMaxScaler**   | Scales features to a fixed range, usually \[0, 1]                           | $x' = \frac{x - x_{\text{min}}}{x_{\text{max}} - x_{\text{min}}}$ | Yes                        | When you want all features to be within the same range, e.g., neural networks |
| **RobustScaler**   | Scales features using statistics that are robust to outliers (median & IQR) | $x' = \frac{x - \text{median}}{\text{IQR}}$, where IQR = Q3 - Q1  | No                         | When data contains many outliers                                              |
