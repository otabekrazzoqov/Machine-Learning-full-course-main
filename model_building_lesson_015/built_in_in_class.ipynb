```python
import plotly.express as px
df = px.data.gapminder().query("year == 2007").query("continent == 'Europe'")
df.loc[df['pop'] < 2.e6, 'country'] = 'Other countries' # Represent only large countries
fig = px.pie(df, values='pop', names='country', title='Population of European continent')
fig.show()
```

```python
import pandas as pd
df=pd.read_csv('heart_disease_uci.csv')
```

```python
import plotly.express as px
import pandas as pd

num_by_gender = df.groupby("gender", as_index=False)["num"].sum()


num_by_gender.loc[num_by_gender['num'] < 50, 'gender'] = 'Other'


fig = px.pie(num_by_gender, values='num', names='gender', 
             title="Sum of 'num' by Gender",color_discrete_sequence=px.colors.sequential.RdBu)
fig.show()
```
