```python
x = df.drop("tax_percentage", axis=1)
y = df["tax_percentage"]
```

## **Linear Regression**

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

lr = LinearRegression()

lr.fit(x_train, y_train)

y_pred = lr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)
```

## **Decision Tree Regressor**

```python
from sklearn.tree import DecisionTreeRegressor

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

dtr = DecisionTreeRegressor()

dtr.fit(x_train, y_train)

y_pred = dtr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)

```

## **Random Forest Regressor**

```python
from sklearn.ensemble import RandomForestRegressor

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

rfr = RandomForestRegressor()

rfr.fit(x_train, y_train)

y_pred = rfr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)

```

## **SVM**

```python
from sklearn.svm import SVR

x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)

svr = SVR()

svr.fit(x_train, y_train)

y_pred = svr.predict(x_test)

mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test, y_pred)

print(mae)
print(mse)
print(rmse)
print(r2)

```

```python
results = [
    ["Linear Regression", 11.73, 207.85, 14.41, 0.33],
    ["Decision Tree Regressor", 14.47, 343.95, 18.54, -0.14],
    ["Random Forest Regressor", 11.18, 187.36, 13.67, 0.377],
    ["Supprt Vector Regressor", 13.16, 276.39, 16.62, 0.08]
]

headers = ["Algorithm", "MAE", "MSE", "RMSE", "r2_score"]

table = tabulate(results, headers=headers, tablefmt = 'grid', floatfmt='.2f')
print(table)
```

## **Ordinal Encoding**

```python
from sklearn.preprocessing import OrdinalEncoder

oe = OrdinalEncoder(categories=[
    ["Low", "Middle", "High"],                # income_level
    ["Low", "Medium", "High", "Very High"]    # subsidy_level
])

df[["income_level", "subsidy_level"]] = oe.fit_transform(
    df[["income_level", "subsidy_level"]]
)
```

## **Label Encoding**


```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()

cols = ["date", "country", "region"]
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
df.sample(10)
```
