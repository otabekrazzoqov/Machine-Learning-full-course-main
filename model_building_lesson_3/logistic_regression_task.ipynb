```python
import pandas as pd
```

```python
df = pd.read_csv("data.csv")
```

```python
df.drop(columns=["Unnamed: 32"], axis=1, inplace = True)
```

    There is only one "object" column in this dataset and it is target column.

    I will use ordinal encoding where i use map() function 

    property_mapping={
            'Apartment':0,
            'Condo':1,
            'Townhouse':2,
            'Single Family':3,
            'Multi-Family':4
        }

        df['property_ordinal'] = df['Property Type'].map(property_mapping)       # ordinal_encoding

```python
diagnosis_mapping = {
    "M" : 1,
    "B" : 0
}

df['diagnosis'] = df["diagnosis"].map(diagnosis_mapping) 
```

```python
df.head()
```

    The dataset contains no missing values and all the columns contains numerical values which means that 
    there is no need for handling missing values and encoding.

```python
numerical_col = df.select_dtypes(include = ["float64", "int64", "int32"]).columns.drop("diagnosis")
```

```python
numerical_col
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

df[numerical_col] = scaler.fit_transform(df[numerical_col])
```

```python
df.head()
```

```python
x = df.drop("diagnosis", axis=1)
y = df["diagnosis"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
y_pred = lr.predict(x_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
c_m = confusion_matrix(y_test, y_pred)
c_m
```

```python
class_report = classification_report(y_test, y_pred)
print(class_report)
```
