#### Handling missing values

```python
def filling_missing_values(df):
    for col in df.columns:
        if df[col].dtype == "object":
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace = True)
    return df
```

```python
df = filling_missing_values(df)
```

```python
df.isnull().sum()
```

    There is no need to encode columns as dtype of columns are already int or float

#### Scaling 

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
def scale(df):
    scaler = MinMaxScaler()
    for col in df.columns:
        df[col] = scaler.fit_transform(df[[col]])
    return df
```

```python
df = scale(df)
```

```python
df.head()
```

    Model Building

```python
from sklearn.linear_model import LogisticRegression
```

```python
x = df.drop("TenYearCHD", axis=1)
y = df['TenYearCHD']
```

    Training

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train, y_train)
```

```python
prediction = model.predict(x_test)
prediction
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
c_m = confusion_matrix(y_test, prediction)
c_m
```

```python
c_rep = classification_report(y_test, prediction)
c_rep
```

```python
print("Confusion Matrix:\n", confusion_matrix(y_test, prediction))
print("Classification Report:\n", classification_report(y_test, prediction))
```
