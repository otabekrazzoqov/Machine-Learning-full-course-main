#### writing a function that handles missing values in a dataset

```python
def missing_values(df):
    for col in df.columns:
        if df[col].isnull().any():
            if df[col].dtype != 'object':
                df[col] = df[col].fillna(df[col].mean(), inplace=True)
            else:
                df[col] = df[col].fillna(df[col].mode()[0], inplace=True)
    return df
```

```python
missing_values(df)
df.info()
```

#### Writing a function that deals with encoding

```python
encoder = LabelEncoder()
```

```python
def encoding(df):
    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].nunique() <=3:
                dummies = pd.get_dummies(df[col], prefix='col', dtype=int)
                df = pd.concat([df.drop(columns=[col]), dummies], axis=1)
            else:
                df[col] = encoder.fit_transform(df[col])
    return df
```

```python
encoding(df)
```

#### Writing a function that scales given dataset.

```python
def scaling(df):
    for col in df.columns:
        if df[col].dtype != "object":
            scaler = MinMaxScaler()
            scaler
            df[col] = scaler.fit_transform(df[[col]])
    return df
```

```python
preprocessed_df = missing_values(scaling(encoding(df)))
preprocessed_df.info()

```
