## Filling missing values

```python
for col in df.columns:
    if df[col].isnull().any():
        if df[col].dtype == 'object':
            df[col].fillna(df[col].mode()[0], inplace=True)
        else:
            df[col].fillna(df[col].mean(), inplace=True)
```

```python
df.info()
```

# Encoding

```python
df['Gender'].value_counts()
```

```python
df = pd.get_dummies(df, columns=['Gender'], prefix='gender', dtype=int)
```

```python
df.info()
```

```python
df['Primary_Device'].value_counts()
```

```python
df['Primary_Device_freq'] = df['Primary_Device'].map(df['Primary_Device'].value_counts())
df.drop(columns='Primary_Device', axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Exceeded_Recommended_Limit'].value_counts()
```

```python
df['Exceeded_Recommended_Limit'] = df['Exceeded_Recommended_Limit'].astype(int)
```

```python
df.info()
```

```python
df['Health_Impacts'].value_counts()
```

```python
df['Health_Impacts_freq'] = df['Health_Impacts'].map(df['Health_Impacts'].value_counts())
df.drop(columns='Health_Impacts', axis=1, inplace=True)
```

```python
df.info()
```

```python
df['Urban_or_Rural'].value_counts()
```

```python
custom_mapping_dict = {
    'Urban':1,
    'Rural':0
}

df['Urban_or_Rural_ordinal'] = df['Urban_or_Rural'].map(custom_mapping_dict)


df.drop(columns='Urban_or_Rural', axis=1, inplace=True)
```

# Scaling

```python
from sklearn.preprocessing import StandardScaler, RobustScaler, MinMaxScaler, LabelEncoder
```

```python
num_col=df.select_dtypes(include=['int64','float64','int32']).columns.tolist()
```

```python
minmax = MinMaxScaler()
```

# Model Training

```python
# Input columns

x = df.drop('Urban_or_Rural_ordinal', axis=1)

x
```

```python
# Output(target) column

y = df['Urban_or_Rural_ordinal'].astype(int)

y
```

```python
# splitting data into train and test

from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split


x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)

model=DecisionTreeClassifier()
```

```python
model
```

```python
model.fit(x_train, y_train)
```

```python
# Prediction 

y_prediction = model.predict(x_test)
```

```python
y_prediction
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test, y_prediction)
```
