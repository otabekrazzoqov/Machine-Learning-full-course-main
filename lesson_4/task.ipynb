```python
df = pd.read_csv('world_bank_data_2025.csv')
```

```python
df.info()
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df['Inflation (CPI %)'].fillna(df['Inflation (CPI %)'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['GDP (Current USD)'].fillna(df['GDP (Current USD)'].mean(),inplace=True)
df['GDP per Capita (Current USD)'].fillna(df['GDP per Capita (Current USD)'].mean(),inplace=True)
df['Unemployment Rate (%)'].fillna(df['Unemployment Rate (%)'].mean(),inplace=True)
df['Interest Rate (Real, %)'].fillna(df['Interest Rate (Real, %)'].mode()[0],inplace=True)
df['Inflation (GDP Deflator, %)'].fillna(df['Inflation (GDP Deflator, %)'].mode()[0],inplace=True)
df['GDP Growth (% Annual)'].fillna(df['GDP Growth (% Annual)'].mode()[0],inplace=True)
df['Current Account Balance (% GDP)'].fillna(df['Current Account Balance (% GDP)'].mean(),inplace=True)
df['Government Expense (% of GDP)'].fillna(df['Government Expense (% of GDP)'].mean(),inplace=True)
df['Government Revenue (% of GDP)'].fillna(df['Government Revenue (% of GDP)'].mean(),inplace=True)
df['Tax Revenue (% of GDP)'].fillna(df['Tax Revenue (% of GDP)'].median(),inplace=True)
df['Gross National Income (USD)'].fillna(df['Gross National Income (USD)'].mean(),inplace=True)
df['Public Debt (% of GDP)'].fillna(df['Public Debt (% of GDP)'].median(),inplace=True)
df['Public Debt (% of GDP)'].fillna(df['Public Debt (% of GDP)'].mean(),inplace=True)
```
